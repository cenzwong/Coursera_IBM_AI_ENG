# Part A: Build a baseline model

```python
#determing the compressive strength of a given concrete sample based on its ingredients
```

The dataset is about the compressive strength of different samples of concrete based on the volumes of the different ingredients that were used to make them. Ingredients include:

1. Cement

2. Blast Furnace Slag

3. Fly Ash

4. Water

5. Superplasticizer

6. Coarse Aggregate

7. Fine Aggregate

```python
#importing the pandas and the Numpy libraries.
import pandas as pd
import numpy as np
```

```python
#download the data and read it into a pandas dataframe.
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
#check how many data points we have.
concrete_data.shape
```

```python
concrete_data.describe()
```

```python
concrete_data.isnull().sum()
```

```python
#Split data into predictors and target
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

```python
predictors.head()
```

```python
target.head()
```

```python
n_cols = predictors.shape[1]
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# build the model
model = regression_model()
```

```python
# fit the model
#we will train and test the model at the same time using the fit method. We will leave out 30% of the data for validation and we will train the model for 100 epochs.
loss_list = np.array([])

for i in range(1,51):
    model.fit(predictors, target, validation_split=0.3, epochs=50, verbose=0)
    results = model.evaluate(predictors, target)
    print(results)
    loss_list = np.append(loss_list, results)

print(loss_list)
```

```python
import statistics as stat 

print("The mean of the MSEs is", np.mean(loss_list))
print("The standard deviation of the MSEs is", stat.stdev(loss_list))
```

# Part B:  Normalize the data

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
n_cols_norm = predictors_norm.shape[1] # number of predictors
```

```python

def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols_norm,)))
    model.add(Dense(1))
    
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model_norm = regression_model()
loss_list_norm = np.array([])

for i in range(1,51):
    model_norm.fit(predictors_norm, target, validation_split=0.3, epochs=50, verbose=0)
    results_norm = model_norm.evaluate(predictors_norm, target)
    print(results_norm)
    loss_list_norm = np.append(loss_list_norm, results_norm)

print(loss_list_norm)
```

```python
## Calculate Mean and Standard Deviation of MSEs

print("The mean of the MSEs is", np.mean(loss_list_norm))
print("The standard deviation of the MSEs is", stat.stdev(loss_list_norm))
print("The mean of the MSEs improved by", np.mean(loss_list)-np.mean(loss_list_norm), "which is a", 100-np.mean(loss_list_norm)/np.mean(loss_list)*100, "percent improvement.")
```

# Part D: Increase the number of hidden layers

```python
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols_norm,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model3_norm = regression_model()
loss_list_norm3 = np.array([])

for i in range(1,51):
    model3_norm.fit(predictors_norm, target, validation_split=0.3, epochs=50, verbose=0)
    results_norm3 = model3_norm.evaluate(predictors_norm, target)
    print(results_norm3)
    loss_list_norm3 = np.append(loss_list_norm3, results_norm3)

print(loss_list_norm3)
```

```python
## Calculate Mean and Standard Deviation of MSEs

print("The mean of the MSEs is", np.mean(loss_list_norm3))
print("The standard deviation of the MSEs is", stat.stdev(loss_list_norm3))
print("The mean of the MSEs improved by", np.mean(loss_list_norm)-np.mean(loss_list_norm3), "which is a", 100-np.mean(loss_list_norm3)/np.mean(loss_list_norm)*100, "percent improvement.")
```
